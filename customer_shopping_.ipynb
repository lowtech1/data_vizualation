**Doanh thu theo danh mục sản phẩm**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

data = pd.read_csv("/content/customer_shopping_data.csv")

data['revenue'] = data['quantity'] * data['price']

# Nhóm dữ liệu theo danh mục và tính tổng doanh thu, sắp xếp giảm dần
revenue_by_category = data.groupby('category')['revenue'].sum().sort_values(ascending=False)

# Vẽ biểu đồ cột
plt.figure(figsize=(12, 6))
sns.barplot(x=revenue_by_category.index,
            y=revenue_by_category.values,
            hue=revenue_by_category.index,
            palette="pastel",
            legend=False)
# Định dạng trục y
def millions(x, pos):
    return f'{int(x/1e6):,}M'
formatter = FuncFormatter(millions)
plt.gca().yaxis.set_major_formatter(formatter)

plt.title('Doanh thu theo danh mục sản phẩm', fontsize=16)
plt.xlabel('Danh mục', fontsize=12)
plt.ylabel('Doanh thu', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.show()
```

**Số lượng đơn hàng theo ngày của các trung tâm thương mại**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("/content/customer_shopping_data.csv")

data['day_of_week'] = pd.to_datetime(data['invoice_date'], format='%d/%m/%Y').dt.day_name()
# Tính tổng số lượng đơn hàng theo ngày và trung tâm thương mại
sales_by_day_mall = data.groupby(['day_of_week', 'shopping_mall'])['quantity'].sum().reset_index()

# Vẽ biểu đồ nhiệt(heatmap)
plt.figure(figsize=(12,6))
sns.heatmap(sales_by_day_mall.pivot(index='day_of_week',
                                    columns='shopping_mall',
                                    values='quantity'),
                                    annot=True, fmt='g')

plt.title('Số lượng đơn hàng theo ngày của các trung tâm thương mại')
plt.xlabel('Trung tâm thương mại')
plt.ylabel('Ngày trong tuần')
plt.show()
```

**Thể hiện doanh thu theo thời gian**

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Giả sử df đã được định nghĩa và chứa cột 'price', 'quantity', và 'invoice_date'

# Tính toán doanh thu
df['revenue'] = df['price'] * df['quantity']

# Chuyển đổi cột 'invoice_date' sang kiểu datetime
df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True)

# Tính tổng doanh thu theo tháng
revenue_by_date = df.groupby(df['invoice_date'].dt.to_period('M'))['revenue'].sum()

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))

# Chuyển đổi index của revenue_by_date thành chuỗi để matplotlib có thể xử lý
x_values = [str(period) for period in revenue_by_date.index]

# Vẽ từng đoạn với màu sắc khác nhau
for i in range(1, len(revenue_by_date)):
    color = 'g' if revenue_by_date.iloc[i] > revenue_by_date.iloc[i - 1] else 'r'
    plt.plot([x_values[i - 1], x_values[i]],
             [revenue_by_date.iloc[i - 1], revenue_by_date.iloc[i]], color=color)

plt.title('Tổng doanh thu theo thời gian', fontsize=16)
plt.xlabel('Thời gian', fontsize=12)
plt.ylabel('Tổng doanh thu', fontsize=12)

# Định dạng trục y
def currency_format(x, _):
    return f'{x:,.0f}'

plt.gca().yaxis.set_major_formatter(FuncFormatter(currency_format))
plt.ylim(0, revenue_by_date.max() * 1.5)
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA



df = pd.read_csv(r'/content/customer_shopping_data.csv')


# Convert invoice_date to datetime
df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%d/%m/%Y')
df['month'] = df['invoice_date'].dt.to_period('M')
```

**Phân cụm khách hàng**

```python
from sklearn.preprocessing import StandardScaler
# Tạo tính năng RFM
current_date = df['invoice_date'].max()

rfm = df.groupby('customer_id').agg({
    'invoice_date': lambda x: (current_date - x.max()).days,  # Recency
    'invoice_no': 'count',  # Frequency
    'price': 'sum'  # Monetary
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

# Chuẩn bị dữ liệu để phân cụm
features = pd.get_dummies(df, columns=['gender', 'category', 'payment_method', 'shopping_mall'])
features = features.groupby('customer_id').agg({
    'age': 'mean',
    'quantity': 'sum',
    'price': 'sum'
})

# Kết hợp RFM và các tính năng khác
features = features.join(rfm)

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Giảm chiều dữ liệu bằng PCA
pca = PCA(n_components=0.95)  # Giữ 95% phương sai
features_pca = pca.fit_transform(features_scaled)

# Xác định số lượng cụm tối ưu bằng phương pháp Elbow
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(features_pca)
    wcss.append(kmeans.inertia_)

# Vẽ đồ thị Elbow
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Phân cụm bằng K-means
optimal_clusters = 4  # Giả sử chúng ta chọn 4 cụm sau khi xem xét đồ thị Elbow
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = kmeans.fit_predict(features_pca)

# Thêm nhãn cụm vào dữ liệu gốc
features['Cluster'] = cluster_labels

# Phân tích các cụm
cluster_analysis = features.groupby('Cluster').mean()
print(cluster_analysis)

# Vẽ biểu đồ phân bố các cụm
plt.figure(figsize=(10, 8))
for i in range(optimal_clusters):
    cluster_data = features_pca[cluster_labels == i]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {i}')

plt.title('Customer Segments')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()

# In ra một số thông tin về các cụm
for i in range(optimal_clusters):
    print(f"\nCluster {i}:")
    cluster_customers = features[features['Cluster'] == i]
    print(f"Number of customers: {len(cluster_customers)}")
    print(f"Average Recency: {cluster_customers['Recency'].mean():.2f} days")
    print(f"Average Frequency: {cluster_customers['Frequency'].mean():.2f} purchases")
    print(f"Average Monetary: ${cluster_customers['Monetary'].mean():.2f}")
    print(f"Average Age: {cluster_customers['age'].mean():.2f} years")
```
